```python
medida=float(input('Digite uma medida em metros:'))

km=medida*0.001
hm=medida*0.01
dam=medida*0.1
cm=medida*100
dm=medida*10
mm=medida*1000

print('Medida em km:',km)
print('Medida em hm:',hm)
print('Medida em dam:',dam)
print('Medida em dm',dm)
print('Medida em cm:',cm)
print('Medida em Mm:',mm)

```
